## 2.4 Columns Requiring Special Interpretation

**Wrong data type (needs cleaning before analysis):**
- `price` — stored as text with '$' symbol (e.g. "$113.97"). Must remove '$' and convert to number before any price analysis.
- `bathrooms_text` — stored as text like "1 shared bath", "1 private bath". Must extract the number from the text.

**Entirely missing — unusable in this dataset:**
- `instant_bookable` — 100% NaN. Should be 't' or 'f' but entire column is missing. Cannot use this column.
- `host_response_rate` — 100% NaN. Cannot analyse host responsiveness.
- `host_acceptance_rate` — 100% NaN. Cannot analyse host selectivity.
- `host_since` — 100% NaN. Cannot calculate how long hosts have been on the platform.
- `neighbourhood` — 100% NaN. Use `neighbourhood_cleansed` instead.
- `calendar_updated` — 100% NaN. Unusable.

**Partially missing — use with caution:**
- `license` — 82.5% NaN. NYC law requires short-term rental registration, so NaN likely means the listing is non-compliant.
- `bedrooms` — 36.3% missing. Cannot rely on this for bedroom-based analysis without imputation.
- `bathrooms` — 33.4% missing. Same issue.
- `price` — 28.9% NaN. Nearly 1 in 3 listings has no price set.
- `review_scores_*` — all around 28.3% NaN. Listings with no reviews have no scores.

**Non-obvious meaning:**
- `available` (calendar file) — values are 't' or 'f' text strings, not True/False. Also, 'f' means unavailable but could be either booked by a guest OR blocked by the host — cannot tell the difference.
- `reviews_per_month` — not from Airbnb. Calculated by Inside Airbnb as total reviews divided by months since first review.
- `calculated_host_listings_count` — reflects the host's listing count at scrape time only, not historically.
- `neighbourhood_cleansed` — standardized version of neighbourhood by Inside Airbnb. Always use this instead of the raw `neighbourhood` column.

## 2.5 Dataset Limitations

**1. Many listings have no price or reviews**
- 8,744 listings (29% of all listings) have no price set.
  This means almost 1 in 3 listings cannot be used in price analysis.
- 8,559 listings have zero reviews.
  These listings may be new or inactive, so we cannot measure their demand.
- All listings have coordinates, so location data is complete.

**2. Scraping failures — 12 columns are completely empty**
The following columns have 0% data — the scraper failed to collect them:
host_since, host_response_time, host_response_rate, host_acceptance_rate,
instant_bookable, neighbourhood, calendar_updated, host_verifications,
host_thumbnail_url, host_neighbourhood, host_total_listings_count,
neighborhood_overview.
These columns cannot be used in any analysis at all.

**3. Orphaned records across files**
- 296 IDs exist in calendar but have no matching listing.
- 239 IDs exist in reviews but have no matching listing.
These are listings that were probably deleted from Airbnb after
the calendar and reviews were scraped but before listings.csv was collected.

**4. No duplicates found**
- 0 duplicate listing IDs
- 0 duplicate calendar rows
- 0 duplicate review IDs
This is good — the data is clean in terms of duplication.

**5. Thin neighbourhoods — some areas have very few listings**
57 neighbourhoods have fewer than 10 listings.
The worst examples are Fort Wadsworth, Willowbrook, Oakwood and Dongan Hills
with only 1 listing each. You cannot draw any conclusions about a neighbourhood
with 1 or 2 listings — the data is too thin to be meaningful.

**6. Host data is almost entirely missing**
- host_since: missing for all 30,259 listings
- host_response_rate: missing for all 30,259 listings  
- host_acceptance_rate: missing for all 30,259 listings
- host_name and host_is_superhost: missing for 349 listings
We cannot analyse how long hosts have been active or how responsive they are.

**7. Review volume dropped sharply in 2020**
Reviews went from 91,559 in 2019 down to 36,683 in 2020.
This is because of COVID-19 — people stopped travelling and bookings collapsed.
Any trend analysis including 2020 will be distorted by this.
Also 2026 only shows 58,572 reviews because the data was scraped in June 2026,
so the year is not complete yet.

**8. Price outliers exist**
- Minimum price is $4.58 per night
- Maximum price is $30,972.96 per night
- 12 listings are priced above $10,000 per night
These extreme prices are likely errors or luxury properties and will
skew any average price calculation if not handled carefully.

**9. This data is a snapshot, not historical**
The scrape happened on 2026-06-23. The calendar only covers
2026-06-14 to 2027-06-22 — one year forward from scrape date.
There is no data about how prices or availability changed in the past.
We can only see what the market looks like at this one moment in time.

## 2.7 Business Domain Context

This dataset represents the New York City short-term rental market captured
by Inside Airbnb on 2026-06-23.

### Listing
A listing is a property made available for short-term rental on Airbnb.
It is the **core entity** of this entire dataset, every other file
connects back to it through a unique listing ID.

A listing can be one of three types:
- **Entire home/apartment** — guest has the whole property to themselves
- **Private room** — guest has their own room but shares common areas
- **Shared room** — guest shares a sleeping space with others

Key facts from the output:
- **30,259 listings** across all five boroughs of New York City
- Each listing has up to **90 columns** of information including price,
  location, room type, bedrooms, bathrooms, amenities and review scores
- Prices range from **$4.58 to $30,972.96** per night
- **8,744 listings (28.9%) have no price** — excluded from price analysis
- **8,559 listings have zero reviews** — demand unknown for these listings
- **12 listings are priced above $10,000** — likely errors or luxury properties


### Host
A host is the person or company that owns and manages one or more listings.
Hosts are the **supply side** of the Airbnb marketplace, they provide
the accommodation that guests book and pay for.

Key facts from the output i got:
- **16,474 unique hosts** managing 30,259 listings
- Average of **1.84 listings per host** — but this is unequal in reality.
  Some hosts manage just one listing casually while others run many
  listings as a full commercial business
- Host data is **embedded inside the listings file** —
  there is no separate hosts table in this dataset

> **Critical Data Gap** — This scrape has severe host data issues.
> `host_since`, `host_response_rate` and `host_acceptance_rate`
> are **100% missing** across all 30,259 listings.
> `host_name` and `host_is_superhost` are missing for **349 listings**.
> This means host-level analysis is very limited in this dataset.


### Review
A review is written feedback left by a guest after completing a stay.
Reviews are one of the **most valuable entities** in this dataset because
Airbnb does not share real booking or transaction data publicly.
This means **review count is our only proxy for booking demand**.

Key facts from our data:
- **990,170 reviews** spanning from 2009-05-25 to 2026-06-22
- More reviews generally means more bookings — but actual demand
  is always higher because **not every guest leaves a review**
